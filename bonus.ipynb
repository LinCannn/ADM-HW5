```python

from auxiliary_files.ConnectedComponents import *
def compare_implementations_with_timing(csv_path: str, start_date: str, end_date: str):
    print("Running UnionFind Implementation...")
    unionfind_result = find_connected_components_with_timing(csv_path, start_date, end_date)

    print("\nRunning GraphFrames Implementation...")
    graphframes_result = find_connected_components_graph_with_timing(csv_path, start_date, end_date)

    print("\nTiming Analysis:")
    print("-" * 80)
    print(f"{'Section':<25} {'UnionFind (s)':<15} {'GraphFrames (s)':<15} {'Difference (s)':<15}")
    print("-" * 80)

    sections = ['setup_time', 'data_loading_time', 'processing_time', 'component_analysis_time', 'total_time']

    for section in sections:
        unionfind_time = unionfind_result['timing_metrics'][section]
        graphframes_time = graphframes_result['timing_metrics'][section]
        diff = unionfind_time - graphframes_time
        print(f"{section:<25} {unionfind_time:>14.2f} {graphframes_time:>14.2f} {diff:>14.2f}")

    return unionfind_result, graphframes_result


csv_path = '/content/drive/MyDrive/ADM-HW5/archive_usa_airport/flights_final.csv'
start_date = '1990-01-01'
end_date = '2005-12-31'

unionfind_result, graphframes_result = compare_implementations_with_timing(csv_path, start_date, end_date)
```
